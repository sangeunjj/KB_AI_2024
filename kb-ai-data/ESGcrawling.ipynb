```python
import pandas as pd
import openpyxl
from urllib.request import urlopen
from bs4 import BeautifulSoup
from html_table_parser import parser_functions as parser
```

```python
df = pd.DataFrame(columns=['NO', '기업명', '기업코드 첨부내용', 'ESG등급', '환경', '사회', '지배구조', '평가년도', '비고 (조정시기)'])
for page_num in range(1,105):
    url = f'https://www.cgs.or.kr/business/esg_tab04.jsp?pg={page_num}&pp=10&skey=&svalue=&sfyear=2023&styear=2023&sgtype=&sgrade=#ui_contents'
    result = urlopen(url)
    html = result.read()
    soup = BeautifulSoup(html, 'html.parser')
    temp = soup.find_all('table')
    p = parser.make2d(temp[1])
    columns = p[0]
    df2=pd.DataFrame(p[1:],columns=p[0])
    df = pd.concat([df,df2],axis=0)
df
```

```python
df=df.rename(columns={'기업코드 첨부내용':'기업코드'})
df=df[['기업명','기업코드','ESG등급','환경','사회','지배구조','평가년도']]
df
```

```python
df.to_csv('ESG등급.csv',encoding='cp949')
```
