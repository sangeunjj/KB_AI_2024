### 1-2. 데이터 형태 변환

```python
# 영문 변수는 기업 개황 데이터 -> 필요컬럼만 한글명으로 변환

df = df.rename(columns={
    'induty_code': '업종코드',
    'est_dt': '설립일',
    'ceo_nm': '대표자명',
    'jurir_no': '법인등록번호',
    'bizr_no': '사업자등록번호',
    'adres': '주소'
})
```

```python
#날짜 데이터 datetime 형식으로 변환
df['설립일'] = pd.to_numeric(df['설립일'], errors='coerce').astype('Int64')
df['설립일'] = pd.to_datetime(df['설립일'], format='%Y%m%d', errors='coerce')
```

```python
#object 데이터 범주화, 서열화

# 순서를 지정하여 범주형 데이터로 변환
esg_order = ['D', 'C', 'B', 'B+', 'A', 'A+']
df['ESG등급'] = pd.Categorical(df['ESG등급'], categories=esg_order, ordered=True)
df['환경'] = pd.Categorical(df['환경'], categories=esg_order, ordered=True)
df['사회'] = pd.Categorical(df['사회'], categories=esg_order, ordered=True)
df['지배구조'] = pd.Categorical(df['지배구조'], categories=esg_order, ordered=True)
```

# 2. 파생변수 생성

```python
#미수금 대비 현금성 자산 충분성 분석 = (현금및현금성자산 / 미수금) 비율
periods = ['당액', '전기', '전전기']
for period in periods:
    df[f'미수금대현금자산비율_{period}'] = df[f'미수금_{period}'] / df[f'현금및현금성자산_{period}']
```

```python
#금융비용대매출액비율변화율 = (매출액/금융비용) 비율
for period in periods:
    df[f'금융비용대매출액비율_{period}'] = df[f'매출액_{period}'] / df[f'금융비용_{period}']
```

```python
# 1. 영업활동현금흐름 변화율 계산
  # 전전기 -> 전기 변화율 계산
df['영업활동현금흐름_전기_변화율'] = ((df['영업활동현금흐름_전기'] - df['영업활동현금흐름_전전기']) / df['영업활동현금흐름_전전기']) * 100
  # 전기 -> 당액 변화율 계산
df['영업활동현금흐름_당액_변화율'] = ((df['영업활동현금흐름_당액'] - df['영업활동현금흐름_전기']) / df['영업활동현금흐름_전기']) * 100

# 2. 투자활동현금흐름 변화율 계산
df['투자활동현금흐름_전기_변화율'] = ((df['투자활동현금흐름_전기'] - df['투자활동현금흐름_전전기']) / df['투자활동현금흐름_전전기']) * 100
df['투자활동현금흐름_당액_변화율'] = ((df['투자활동현금흐름_당액'] - df['투자활동현금흐름_전기']) / df['투자활동현금흐름_전기']) * 100

# 3. 재무활동현금흐름 변화율 계산
df['재무활동현금흐름_전기_변화율'] = ((df['재무활동현금흐름_전기'] - df['재무활동현금흐름_전전기']) / df['재무활동현금흐름_전전기']) * 100
df['재무활동현금흐름_당액_변화율'] = ((df['재무활동현금흐름_당액'] - df['재무활동현금흐름_전기']) / df['재무활동현금흐름_전기']) * 100
```

```python
### 2. 부채 및 자본 구조 관련 파생변수 ###
periods = ['당액', '전기', '전전기']

for period in periods:
    # 2.1. 부채비율 (Debt to Assets Ratio)
    df[f'부채비율_{period}'] = df[f'부채총계_{period}'] / df[f'자산총계_{period}']

    # 2.2. 유동비율 (Current Ratio)
    df[f'유동비율_{period}'] = df[f'유동자산_{period}'] / df[f'유동부채_{period}']

    # 2.3. 영업이익률 (Operating Profit Margin)
    df[f'영업이익률_{period}'] = df[f'영업이익_{period}'] / df[f'매출액_{period}']

    # 2.4. 금융비용 대 영업이익 비율 (Interest Expense to Operating Profit Ratio)
    df[f'금융비용대영업이익비율_{period}'] = df[f'금융비용_{period}'] / df[f'영업이익_{period}']

    # 2.5. 자본 대비 충당부채 비율 (Provisions to Equity Ratio)
    df[f'자본대비충당부채비율_{period}'] = df[f'충당부채_{period}'] / df[f'자본금_{period}']
```

```python
### 3. 주요 재무지표 산출 ###
# 1. 영업이익률 변화율 계산
df['영업이익률_전기_변화율'] = ((df['영업이익률_전기'] - df['영업이익률_전전기']) / df['영업이익률_전전기']) * 100
df['영업이익률_당액_변화율'] = ((df['영업이익률_당액'] - df['영업이익률_전기']) / df['영업이익률_전기']) * 100

# 2. 부채비율 변화율 계산
df['부채비율_전기_변화율'] = ((df['부채비율_전기'] - df['부채비율_전전기']) / df['부채비율_전전기']) * 100
df['부채비율_당액_변화율'] = ((df['부채비율_당액'] - df['부채비율_전기']) / df['부채비율_전기']) * 100

# 3. 유동비율 변화율 계산
df['유동비율_전기_변화율'] = ((df['유동비율_전기'] - df['유동비율_전전기']) / df['유동비율_전전기']) * 100
df['유동비율_당액_변화율'] = ((df['유동비율_당액'] - df['유동비율_전기']) / df['유동비율_전기']) * 100
```

```python
### 설립일자로부터 오늘까지 경과일수 생성
from datetime import datetime
# 설립일자로부터 오늘(24.8.11)까지 일수
today = pd.to_datetime(datetime.today().date())
df['오늘까지_일수'] = (today - df['설립일']).dt.days
```
